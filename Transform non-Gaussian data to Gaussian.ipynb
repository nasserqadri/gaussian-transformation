### Transform non-Gaussian data to Gaussian

Transforming non-gaussian (brownian noise) data to Gaussian, but maintaining monotonic pattern

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
n = 500
x = np.cumsum(np.random.randn(n))
y = (stats.rankdata(x)/(n+1))*2 -1
y = np.arctanh(y)
print(np.min(y), np.max(y))
```

```python
fig, ax = plt.subplots(2,2 , figsize=(8,8))

ax[0,0].plot(x)
ax[0,0].set_title('Original data')

ax[0,1].plot(y)
ax[0,1].set_title('Transformed data')

ax[1,0].hist(x, bins=40)
ax[1,0].set_title('Original data')

ax[1,1].hist(y, bins=40)
ax[1,1].set_title('Transformed data')



plt.show()
```

```python
plt.plot(x,y,'s')
plt.xlabel('Original')
plt.ylabel('Transformed')
plt.show()
```
